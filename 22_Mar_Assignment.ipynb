### 1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

To calculate the Pearson correlation coefficient between two variables, you need to first compute the covariance and the standard deviations of the two variables. Let's assume that you have collected the following data:

Student	Study Time (hours)	Exam Score

1	4	85

2	6	90

3	3	80

4	5	89

5	7	92

To calculate the Pearson correlation coefficient between study time and exam score, you can use the following formula:

r = cov(x, y) / (std(x) * std(y))

where cov(x,y) is the covariance of the two variables, and std(x) and std(y) are their respective standard deviations.

Using the formula, we can compute the correlation coefficient:

cov(x, y) = ((4-5.0)(85-88.4) + (6-5.0)(90-88.4) + (3-5.0)(80-88.4) + (5-5.0)(89-88.4) + (7-5.0)*(92-88.4)) / 5 = 6.8

std(x) = sqrt(((4-5.0)^2 + (6-5.0)^2 + (3-5.0)^2 + (5-5.0)^2 + (7-5.0)^2) / 4) = 1.581

std(y) = sqrt(((85-88.4)^2 + (90-88.4)^2 + (80-88.4)^2 + (89-88.4)^2 + (92-88.4)^2) / 4) = 4.129

r = 6.8 / (1.581 * 4.129) = 0.407

The Pearson correlation coefficient between study time and exam score is 0.407, which indicates a moderate positive linear relationship between the two variables. This means that as the amount of time students spend studying for an exam increases, their final exam scores tend to increase as well, but not perfectly. However, correlation does not imply causation and other factors may also be at play.

### 2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Calculate the Spearman's rank correlation between these two variables and interpret the result.

ρ = 1 - (6 * ∑d^2) / (n(n^2 - 1))
