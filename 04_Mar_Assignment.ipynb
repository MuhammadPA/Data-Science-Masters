### 1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import plotly.graph_objects as go
import seaborn as sns
titanic = sns.load_dataset('titanic')
fig = go.Figure()
fig.add_trace(go.Scatter(x = titanic.age,y =titanic.fare,mode = 'markers'))
```

### 2.Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px

tips_df = px.data.tips
box_plot = px.box(tips_df, x="day", y="total_bill")
box_plot.show()
```

### 3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import plotly.express as px
tips_df = px.data.tips
histogram = px.histogram(tips_df, x='sex', y='total_bill', color='day', pattern_shape='smoker')
histogram.show()
```

### 4.Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
import plotly.express as px
iris_df = px.data.iris()
scatter_matrix = px.scatter_matrix(iris_df,
                                   dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                                   color="species")
scatter_matrix.show()
```

### 5. What is Distplot? Using Plotly express, plot a distplot.

A distplot is a type of histogram with a smooth curve that shows the distribution of a dataset. It is a visualization tool commonly used to analyze the distribution of continuous variables.

```python
import plotly.express as px
import numpy as np

np.random.seed(123)
x = np.random.normal(size=1000)

fig = px.histogram(x=x, nbins=30, opacity=0.7)

fig.update_layout(
    title="Distplot Example",
    xaxis_title="Values",
    yaxis_title="Count"
)
```
