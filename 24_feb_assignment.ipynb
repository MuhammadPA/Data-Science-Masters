### 2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
import pandas as pd

data = pd.DataFrame({
    'A': [10, 20, 30, 40],
    'B': [100, 200, 300, 400],
    'C': [1000, 2000, 3000, 4000]
})

def reindex_dataframe(df):
    new_index = pd.RangeIndex(start=1, stop=2*len(df), step=2)
    df.index = new_index
    return df
reindexed_df = reindex_dataframe(data)
reindexed_df
```

### 3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console. For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should calculate and print the sum of the first three values, which is 60.

```python
import pandas as pd

def calculate_sum(df):
    first_three_values = df['Values'].iloc[:3]
    total_sum = sum(first_three_values)
    print("The sum of the first three values is:", total_sum)
    
data2 = pd.DataFrame({
    'Values': [10, 20, 30, 40, 50],
    'Values1': [100, 200, 300, 400, 500],
    'Values2': [1000, 2000, 3000, 4000, 5000]
})
calculate_sum(data2)
```

### 4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
import pandas as pd

def add_word_count_column(df):
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))
    return df

data3 = pd.DataFrame({'Text': ['The quick brown fox', 'jumps over', 'the lazy dog']})
count = add_word_count_column(data3)
count
```

### 5. How are DataFrame.size() and DataFrame.shape() different?

The DataFrame.size and DataFrame.shape attributes in Pandas are related to the size and shape of a DataFrame, but they have different meanings and return different values.

The DataFrame.size attribute returns the number of elements in a DataFrame. This is simply the total number of cells or values in the DataFrame, calculated as the product of the number of rows and columns in the DataFrame. For example, if a DataFrame has 5 rows and 4 columns, its size is 20.

On the other hand, the DataFrame.shape attribute returns a tuple that contains the number of rows and columns in the DataFrame, respectively. For example, if a DataFrame has 5 rows and 4 columns, its shape is (5, 4). The shape attribute provides a convenient way to quickly get the dimensions of a DataFrame, which can be useful for understanding the structure and size of a dataset.

### 7. You have a Pandas DataFrame df that contains a column named 'Email' that contains emailaddresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address. The username is the part of the email address that appears before the '@' symbol. For example, if the email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your function should extract the username from each email address and store it in the new 'Username' column.

```python
import pandas as pd

def extract_username(email):
    return email.split('@')[0]

data4 = pd.DataFrame({'Email': ['john.doe@example.com', 'jane.smith@example.com', 'bob.johnson@example.com']})
data4['Username'] = data4['Email'].apply(extract_username)
print(data4)
```

### 8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selectsall rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows. For example, if df contains the following values:
A B C

0 3 5 1

1 8 2 7

2 6 9 4

3 2 3 5

4 9 1 2

```python
import pandas as pd

def select_rows(df):
    selected_df = df[(df['A'] > 5) & (df['B'] < 10)]
    return selected_df

data5 = pd.DataFrame({'A': [3, 8, 6, 2, 9], 'B': [5, 2, 9, 3, 1], 'C': [1, 7, 4, 5, 2]})
selected_df = select_rows(data5)
selected_df
```

### 9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
def calculate_stats(df):
    mean_val = df['Values'].mean()
    median_val = df['Values'].median()
    std_val = df['Values'].std()
    print("Mean value: ", mean_val)
    print("Median value: ", median_val)
    print("Standard deviation: ", std_val)
    
data6 = pd.DataFrame({
    'Values': [10, 20, 30, 40, 50],
    'Values1': [100, 200, 300, 400, 500],
    'Values2': [1000, 2000, 3000, 4000, 5000]
})
result=calculate_stats(data6)
```

### 10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
import pandas as pd
import numpy as np

def calculate_moving_average(df):
    window_size = 7
    ma_column_name = 'MovingAverage'
    df = df.sort_values('Date')
    ma = df['Sales'].rolling(window_size, min_periods=1).mean()
    df[ma_column_name] = ma
    return df

dates = pd.date_range('20220301', periods=10)
sales = np.random.randint(10, 100, size=10)
df = pd.DataFrame({'Date': dates, 'Sales': sales})
new_data = calculate_moving_average(df)
new_data
```

### 11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.

For example, if df contains the following values:

Date

0 2023-01-01

1 2023-01-02

2 2023-01-03

3 2023-01-04

4 2023-01-05

```python
import pandas as pd

def add_weekday_column(df):
    df['Weekday'] = pd.to_datetime(df['Date']).dt.day_name()
    return df

data7 = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})
weekday = add_weekday_column(data7)
weekday
```

### 12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
import pandas as pd
import numpy as np

def select_january_data(df):
    start_date = '2023-01-01'
    end_date = '2023-01-31'
    df['Date'] = pd.to_datetime(df['Date'])
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    selected_df = df[mask]   
    return selected_df

data8 = pd.DataFrame({
   'Date': pd.date_range(start='2023-01-01', end='2023-02-28', freq='D'),
   'Values': np.random.randint(1, 100, size=59)
})
selected_df = select_january_data(data8)
selected_df
```
