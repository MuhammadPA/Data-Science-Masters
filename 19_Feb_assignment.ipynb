### 3. What is App routing in Flask? Why do we use app routes?

In Flask, an app route is a URL pattern that maps to a function in your application. When a user 
visits a specific URL, Flask maps that URL to a function that generates the content for that page. 
App routing is the process of defining these URL patterns and their corresponding functions.

App routes are used in Flask to map URLs to specific functions or views that generate the content 
for those URLs. This allows you to create a dynamic website where different URLs correspond to different 
content, such as pages, blog posts, or articles. By defining app routes, you can also create RESTful APIs 
that allow other applications to interact with your Flask application.

### 4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details: Company Name: ABC Corporation, Location: India, Contact Detail: 999-999-9999

### 5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

```python
'''In Flask, the url_for() function is used for URL building. It generates a URL for a specific view 
function based on the name of the function.'''

from flask import Flask
from flask import url_for
from flask import request

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the index page'

@app.route('/user/<username>')
def profile(username):
    return f'This is the profile page for {username}'

@app.route('/about')
def about():
    return 'This is the about page'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))
        # Output: /

        print(url_for('profile', username='John Doe'))
        # Output: /user/John%20Doe

        print(url_for('about'))
        # Output: /about

'''In this example, we have defined three routes: the '/' route, the '/user/<username>' route, and the 
'/about' route. We can use the url_for() function to generate a URL for any of these routes by passing the 
name of the corresponding view function as an argument.

When we run the Flask application and access the '/' route, 'http://localhost:5000/' is displayed in the
browser. However, when we call url_for('index') in our Python code, it returns '/', which is the URL that 
corresponds to the index view function.

Similarly, when we call url_for('profile', username='John Doe'), it returns '/user/John%20Doe', 
which is the URL for the profile view function with the username 'John Doe'. The %20 in the URL is 
URL-encoded whitespace.

Finally, when we call url_for('about'), it returns '/about', which is the URL for the about view function.'''
```
